```python
# Predicting the price of the diamond based on the features of Carat, Cut, Color, Depth, Table,X,Y,Z
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn import svm, preprocessing
df = pd.read_csv("Dataset4/diamonds/diamonds.csv", index_col=0)
df.head()
```

```python
# Fing the unique values in Cut and Clarity
df["cut"].unique()
df["clarity"].unique()
```

```python
#df["cut"].astype("category").cat.codes
```

```python
# Create dictonary for the catagory, Color, Cut to preserve the identiy like good is better than fair
# Creating the dictonary for findng the dummay variables for the string 
cut_class_dict = { "Fair": 1, "Good": 2, "Very Good": 3,"Premium": 4,"Ideal": 5}
#clarity_dict = { 'I3' :1, 'I2':2, 'I1':3,'SI2':4, 'SI1':5, 'VS2':6, 'VS1':7, 'VVS2'8, 'VVS1':9, 'IF':10, 'FL':11}
clarity_dict = { "I3": 1, "I2": 2, "I1": 3,"SI2": 4, "SI1": 5, "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10, "FL": 11}
color_dict = { "J" :1, "I":2, "H":3,"G":4,"F":5,"E":6,"D":7}
# Mapping the dictonary variables to the strings
df['cut'] = df['cut'].map(cut_class_dict)
df['clarity'] = df['clarity'].map(clarity_dict)
df['color']= df['color'].map(color_dict)

df.head()
```

```python
# Shuffle the data 
df = sklearn.utils.shuffle(df)
```

```python
# . values for taking only the values so drop index
# Dropping the drop values for the Y values since we are predicting the price
X = df.drop("price", axis=1).values
# Preprocessing the X values
X= preprocessing.scale(X)
# Storing the predicting the values in Y since predicting the price is the main goal of the project
Y = df["price"].values
# Total amount of the test set
test_size = 200
# Selecting the first 800 dataset in the X for training and from Y for the predicting the cost
X_train = X[:-test_size] 
Y_train = Y[:-test_size]
# Selecting the last 200 dataset in the X for testing and Y for accuracy 
X_test = X[-test_size:] 
Y_test = Y[-test_size:]
# Selecting the Machine Learning Algorithm
clf = svm.SVR(kernel = "linear")
# Fitting the Machine Learning classifier for the data
clf.fit(X_train, Y_train)
```

```python
# Predicting the accuracy of the classifier from the testing data
clf.score(X_test, Y_test)
  
```
